Temperatura de los servidores

##### Lucas trabaja en TI y necesita garantizar que la temperatura de una sala de servidores no supere los 25°C.

##### Quiere un programa que reciba la temperatura actual como entrada y, si es necesario, muestre un mensaje de alerta.

```python
# Programa para verificar la temperatura de una sala de servidores

# Solicitar la temperatura actual
temperatura = float(input("Ingresa la temperatura actual de la sala (°C): "))

# Verificar si supera el límite
if temperatura > 25:
    print("⚠️ Alerta: la temperatura supera los 25°C. Revisa el sistema de enfriamiento.")
else:
    print("✅ Temperatura dentro del rango seguro.")
```
